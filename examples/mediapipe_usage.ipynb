```python
"""
F009 실체 치수
* 키: 161.1cm
목뒤높이: 130.5cm
엉덩이높이: 78.1cm
겨드랑높이: 118.4cm
허리높이: 98.6cm
샅높이: 71.6cm
무릎높이: 42.1cm
머리둘레: 55.5cm
목둘레: 32.4cm
* 젖가슴둘레: 91.3cm
* 허리둘레: 75.9cm
배꼽수준허리둘레: 85.5cm
* 엉덩이둘레: 97.8cm
넙다리둘레: 57.8cm
무릎둘레: 37.5cm
장딴지둘레: 37.4cm
종아리최소둘레: 21.5cm
발목둘레: 24.2cm
편위팔둘레: 27.7cm
편팔꿈치둘레: 23.3cm
손목둘레: 15.1cm
위팔길이: 29.3cm
팔길이: 52.1cm
어깨사이너비: 35.4cm
머리수직길이: 21.1cm
얼굴수직길이: 10.9cm
발크기: 23.2cm
발너비: 10.1cm
얼굴너비: 12.6cm
손직선길이: 16.2cm
손바닥직선길이: 9.4cm
손안쪽가쪽직선길이: 7.4cm
몸무게: 59.2kg
체지방율: 28.6%
성별: F
나이: 44
신발굽높이: 6cm

F010 실제치수
* 키: 167.1cm
목뒤높이: 141.5cm
엉덩이높이: 83.6cm
겨드랑높이: 125.8cm
허리높이: 104.9cm
샅높이: 77.8cm
무릎높이: 44.5cm
머리둘레: 54.5cm
목둘레: 30.9cm
* 젖가슴둘레: 86.1cm
* 허리둘레: 65.7cm
* 배꼽수준허리둘레: 70cm
* 엉덩이둘레: 92.2cm
넙다리둘레: 54.3cm
무릎둘레: 33.2cm
장딴지둘레: 34.5cm
종아리최소둘레: 19.5cm
발목둘레: 22.6cm
편위팔둘레: 27.1cm
편팔꿈치둘레: 21.6cm
손목둘레: 14.4cm
위팔길이: 32.5cm
팔길이: 55.9cm
어깨사이너비: 36.4cm
머리수직길이: 21.4cm
얼굴수직길이: 11cm
발크기: 23.2cm
발너비: 9.2cm
얼굴너비: 12.4cm
손직선길이: 17.1cm
손바닥직선길이: 9.6cm
손안쪽가쪽직선길이: 7cm
* 몸무게: 52.3kg
체지방율: 16%
성별: F
나이: 33
신발굽높이: 1.5cm

F009
[실제]
키: 161.1cm
몸무게: 59.2kg
젖가슴둘레: 91.3cm
배꼽수준허리둘레: 85.5cm
엉덩이둘레: 97.8cm

[예측]
키: 161.1cm
몸무게 (kg): 57.1
가슴둘레 (cm): 95.94
허리 둘레 (cm): 87.71
엉덩이 둘레 (cm): 100.26

F010
[실제]
키: 167.1cm
젖가슴둘레: 86.1cm
배꼽수준허리둘레: 70cm
엉덩이둘레: 92.2cm
몸무게: 52.3kg

[예측]
키: 167.1cm
몸무게 (kg): 61.43
가슴둘레 (cm): 99.3
허리 둘레 (cm): 90.79
엉덩이 둘레 (cm): 94.56

[Mediapipe Pose 랜드마크의 목록]
NOSE (0): 코
LEFT_EYE_INNER (1): 왼쪽 눈 안쪽
LEFT_EYE (2): 왼쪽 눈
LEFT_EYE_OUTER (3): 왼쪽 눈 바깥쪽
RIGHT_EYE_INNER (4): 오른쪽 눈 안쪽
RIGHT_EYE (5): 오른쪽 눈
RIGHT_EYE_OUTER (6): 오른쪽 눈 바깥쪽
LEFT_EAR (7): 왼쪽 귀
RIGHT_EAR (8): 오른쪽 귀
MOUTH_LEFT (9): 입 왼쪽
MOUTH_RIGHT (10): 입 오른쪽
LEFT_SHOULDER (11): 왼쪽 어깨
RIGHT_SHOULDER (12): 오른쪽 어깨
LEFT_ELBOW (13): 왼쪽 팔꿈치
RIGHT_ELBOW (14): 오른쪽 팔꿈치
LEFT_WRIST (15): 왼쪽 손목
RIGHT_WRIST (16): 오른쪽 손목
LEFT_PINKY (17): 왼쪽 새끼손가락
RIGHT_PINKY (18): 오른쪽 새끼손가락
LEFT_INDEX (19): 왼쪽 검지손가락
RIGHT_INDEX (20): 오른쪽 검지손가락
LEFT_THUMB (21): 왼쪽 엄지손가락
RIGHT_THUMB (22): 오른쪽 엄지손가락
LEFT_HIP (23): 왼쪽 엉덩이 (고관절)
RIGHT_HIP (24): 오른쪽 엉덩이 (고관절)
LEFT_KNEE (25): 왼쪽 무릎
RIGHT_KNEE (26): 오른쪽 무릎
LEFT_ANKLE (27): 왼쪽 발목
RIGHT_ANKLE (28): 오른쪽 발목
LEFT_HEEL (29): 왼쪽 발뒤꿈치
RIGHT_HEEL (30): 오른쪽 발뒤꿈치
LEFT_FOOT_INDEX (31): 왼쪽 발끝
RIGHT_FOOT_INDEX (32): 오른쪽 발끝
"""

import cv2
import mediapipe as mp
import math
import matplotlib.pyplot as plt


scaling_factor_chest = 3.5   # 가슴둘레 스케일링 계수
scaling_factor_waist = 3.2   # 허리둘레 스케일링 계수
scaling_factor_hip = 6.1     # 엉덩이둘레 스케일링 계수

# 신체 랜드마크 간 거리 계산 함수 (유클리드 거리)
def calculate_distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# 신체 치수 추정 함수
def estimate_body_measurements(height_cm, landmarks):
    # 키를 미터로 변환
    height_m = height_cm / 100
    
    # 스케일링 계수: 현실적인 비율로 조정하기 위해 계수를 적용

    
    # 신체의 주요 랜드마크 추출
    left_shoulder = landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER]
    right_shoulder = landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER]
    left_hip = landmarks[mp_pose.PoseLandmark.LEFT_HIP]
    right_hip = landmarks[mp_pose.PoseLandmark.RIGHT_HIP]
    
    # 어깨 너비 계산 (왼쪽 어깨 ~ 오른쪽 어깨 거리)
    shoulder_width_ratio = calculate_distance(left_shoulder.x, left_shoulder.y, right_shoulder.x, right_shoulder.y)
    
    # 엉덩이 너비 계산 (왼쪽 엉덩이 ~ 오른쪽 엉덩이 거리)
    hip_width_ratio = calculate_distance(left_hip.x, left_hip.y, right_hip.x, right_hip.y)
    
    # 비율 기반 초기 추정 + 스케일링 계수 적용
    chest_circumference = shoulder_width_ratio * height_cm * scaling_factor_chest  # 가슴둘레
    waist_circumference = chest_circumference * (scaling_factor_waist / scaling_factor_chest)  # 허리둘레
    hip_circumference = hip_width_ratio * height_cm * scaling_factor_hip  # 엉덩이 둘레
    
    # BMI 기반 몸무게 추정 (스케일링이 적용된 신체 치수를 사용하여 몸무게를 추정)
    bmi = 22  # 평균적인 BMI 값
    weight_kg = bmi * (height_m ** 2)  # BMI를 통한 몸무게 추정

    return {
        "몸무게 (kg)": round(weight_kg, 2),  # 몸무게는 kg 단위
        "가슴둘레 (cm)": round(chest_circumference, 2),  # 가슴둘레는 cm 단위
        "허리 둘레 (cm)": round(waist_circumference, 2),  # 허리둘레는 cm 단위
        "엉덩이 둘레 (cm)": round(hip_circumference, 2)  # 엉덩이둘레는 cm 단위
    }

# 이미지 파일 경로
IMAGE_PATH = ['../images/01_01_F009_03.jpg',
              '../images/01_01_F009_07.jpg',
              '../images/01_01_F010_04.jpg',
              '../images/01_01_F010_15.jpg']

for image_path in IMAGE_PATH:
    # Mediapipe의 Pose 모듈 초기화
    mp_pose = mp.solutions.pose
    mp_drawing = mp.solutions.drawing_utils

    # OpenCV로 이미지 불러오기
    image = cv2.imread(image_path)

    # BGR 이미지를 RGB로 변환 (Mediapipe는 RGB 이미지를 필요로 함)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Mediapipe Pose 모델을 사용하여 신체 랜드마크 추출
    with mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5) as pose:
        result = pose.process(image_rgb)

        # 신체 랜드마크가 감지되었는지 확인
        if result.pose_landmarks:
            # Mediapipe 랜드마크 그리기
            annotated_image = image.copy()
            mp_drawing.draw_landmarks(annotated_image, result.pose_landmarks, mp_pose.POSE_CONNECTIONS)

            # 신체 랜드마크 좌표 추출
            landmarks = result.pose_landmarks.landmark
            
            # 사용자의 신장 입력
            user_height = float(input("사용자의 신장(cm)을 입력하세요: "))

            # 신체 치수 추정
            measurements = estimate_body_measurements(user_height, landmarks)
            
            # 결과 출력
            print(f'\n{image_path}')
            for key, value in measurements.items():
                print(f"{key}: {value}")
            
            
            # 이미지 표시
            plt.figure(figsize=(10, 10))
            plt.imshow(annotated_image)
            plt.axis('off')  # 축 숨기기
            plt.show()
            
            # cv2.imshow("Pose Landmarks", annotated_image)
            # cv2.waitKey(0)
            # cv2.destroyAllWindows()
        else:
            print("신체 랜드마크를 찾을 수 없습니다.")
```
